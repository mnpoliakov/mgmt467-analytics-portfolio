### Q1 â€” Exploratory vs. Explanatory
**Prompt:** *"Explain the difference between exploratory and explanatory visualization."*  
**Hints:** Consider audience, purpose, level of polish, and how much iteration you expect.

ðŸ“ **Your response:**
- Exploratory visualization is used for discovering insights and patterns in data. It's often iterative, less polished, and primarily for the data analyst's understanding. The audience is typically the data analyst or team exploring the data.
- Explanatory visualization is used to communicate a specific finding or story to an audience. It's polished, clear, and designed to highlight key insights effectively. The audience is typically a broader group, like stakeholders or the public.
- Think of exploratory as searching for a story in the data, while explanatory is telling that story clearly and concisely to others.

### Q2 â€” Chart Types & When to Use Them
**Prompt:** *"Describe 5 common chart types (e.g., bar, line, scatter, histogram, map) and the type of data analysis question each is best suited to answer."*  

For each chart, list: **Question archetype**, **Data shape**, **Pitfalls**, **Alternatives**.

ðŸ“ **Your response (template):**
1. Bar:
Question archetype: Comparison across categories; Ranking.
Data shape: One categorical variable and one quantitative variable.
Pitfalls: Too many categories make it unreadable; Misleading y-axis scale (not starting at zero); Using 3D effects; Poor sorting.
Alternatives: Column chart, Stacked bar chart, Marimekko chart.

2. Line:
Question archetype: Trends over time; Showing change or evolution.
Data shape: One temporal variable and one quantitative variable.
Pitfalls: Too many lines overcrowding the chart; Using lines for non-sequential data; Misleading y-axis scale.
Alternatives: Area chart, Dot plot, Connected scatter plot.

3. Scatter:
Question archetype: Relationship between two quantitative variables; Identifying correlation or clusters.
Data shape: Two quantitative variables.
Pitfalls: Overplotting with too many points; Misinterpreting correlation as causation; Poor choice of axis ranges.
Alternatives: Bubble chart, Heatmap, Density plot.

4. Histogram:
Question archetype: Distribution of a single quantitative variable; Understanding frequency or density.
Data shape: One quantitative variable.
Pitfalls: Choosing an inappropriate number of bins; Misinterpreting as a bar chart for categories.
Alternatives: Density plot, Box plot, Violin plot.

5. Map:
Question archetype: Geographic patterns; Showing spatial distribution.
Data shape: Geographic data (latitude/longitude, regions, etc.) and one or more quantitative variables.
Pitfalls: Misleading color scales; Using inappropriate map projections; Relying solely on choropleth maps for population density issues.
Alternatives: Dot map, Symbol map, Cartogram.

### Create a Synthetic Airline Dataset (if you don't have one)
Use this dataset to prototype visualizations and dashboards. It contains realistic columns for the KPIs you defined. You can always swap in your BigQuery data later.


````python
# Generate a synthetic airline dataset
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(467)

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')
carriers = ['AA', 'DL', 'UA', 'WN', 'AS']
origins = ['ATL','DFW','DEN','ORD','LAX','SEA','SFO','PHX','CLT','MCO']
dests =   ['JFK','LGA','BOS','SFO','LAX','SEA','MIA','IAH','IAD','PHL']

rows = []
for d in dates:
    for carrier in carriers:
        # number of flights per day per carrier
        n = np.random.randint(80, 140)
        for _ in range(n):
            o = np.random.choice(origins)
            de = np.random.choice(dests)
            if o == de:
                continue
            distance = np.random.randint(300, 2800)
            seats = np.random.choice([90,120,150,180,210])
            load_factor = np.clip(np.random.normal(0.82, 0.07), 0.55, 0.98)
            pax = int(seats * load_factor)
            dep_delay = int(np.random.normal(8, 15))
            arr_delay = dep_delay + int(np.random.normal(0, 12))
            dep_delay = max(dep_delay, -5)
            arr_delay = max(arr_delay, -10)
            cancelled = np.random.rand() < 0.025
            diverted = np.random.rand() < 0.005
            revenue = pax * np.random.uniform(120, 420)
            casm = np.random.uniform(0.08, 0.16)  # cost per available seat mile
            cost = casm * seats * distance
            fuel_burn = distance * np.random.uniform(3.0, 4.5)  # gallons per seat-mile proxy
            nps = np.random.normal(35, 15)  # if you want a customer metric proxy
            rows.append({
                'date': d,
                'carrier': carrier,
                'origin': o,
                'dest': de,
                'distance_miles': distance,
                'seats': seats,
                'passengers': pax,
                'dep_delay_min': dep_delay,
                'arr_delay_min': arr_delay,
                'cancelled': int(cancelled),
                'diverted': int(diverted),
                'revenue_usd': round(revenue, 2),
                'cost_usd': round(cost, 2),
                'fuel_burn_gal': round(fuel_burn, 1),
                'nps': round(nps, 1)
            })

df_flights = pd.DataFrame(rows)

# Derived KPIs
df_flights['on_time'] = (df_flights['arr_delay_min'] <= 15).astype(int)
df_flights['load_factor'] = (df_flights['passengers'] / df_flights['seats']).round(3)
df_flights['yield_usd_per_pax_mile'] = (df_flights['revenue_usd'] / (df_flights['passengers'] * df_flights['distance_miles']).clip(lower=1)).round(4)
df_flights['asm'] = df_flights['seats'] * df_flights['distance_miles']  # available seat miles
df_flights['casm'] = (df_flights['cost_usd'] / df_flights['asm']).round(4)
df_flights['month'] = df_flights['date'].dt.to_period('M').astype(str)

df_flights.head()
````

````python
# Prompt E
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create the scatter plot
plt.figure(figsize=(10, 8))
scatter_plot = sns.scatterplot(x='cost_usd', y='revenue_usd', hue='casm', data=df_flights, palette='viridis', alpha=0.6, s=50)

# Add a diagonal y=x reference line
max_val = max(df_flights['cost_usd'].max(), df_flights['revenue_usd'].max())
plt.plot([0, max_val], [0, max_val], color='red', linestyle='--', label='Cost = Revenue')

# Add labels and title
plt.title('Cost vs. Revenue Colored by CASM', fontsize=16)
plt.xlabel('Cost (USD)', fontsize=12)
plt.ylabel('Revenue (USD)', fontsize=12)
plt.legend(title='CASM')
plt.grid(True)

# Use tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
````

````python
# Prompt D
import matplotlib.pyplot as plt
import seaborn as sns

# Compute mean load factor by route (origin-dest pair)
df_flights['route'] = df_flights['origin'] + ' - ' + df_flights['dest']
mean_load_factor_by_route = df_flights.groupby('route')['load_factor'].mean().reset_index()

# Take the top 15 routes by average load factor
top_15_routes = mean_load_factor_by_route.nlargest(15, 'load_factor')

# Sort for plotting
top_15_routes = top_15_routes.sort_values('load_factor', ascending=False)

# Create the horizontal bar chart
plt.figure(figsize=(10, 8))
sns.barplot(x='load_factor', y='route', data=top_15_routes, palette='viridis')

# Add labels and title
plt.title('Top 15 Routes by Average Load Factor', fontsize=16)
plt.xlabel('Average Load Factor', fontsize=12)
plt.ylabel('Route', fontsize=12)

# Use tight layout for better spacing and readable labels
plt.tight_layout()

# Show the plot
plt.show()
````

````python
# Prompt C
import matplotlib.pyplot as plt
import seaborn as sns

# Create the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df_flights['arr_delay_min'], bins=50, kde=False)

# Add a vertical reference line at 0
plt.axvline(0, color='red', linestyle='--', label='0 Minute Delay')

# Calculate and annotate the percent of flights within [-5, 15] minutes
on_time_flights = df_flights[(df_flights['arr_delay_min'] >= -5) & (df_flights['arr_delay_min'] <= 15)]
percent_on_time = (len(on_time_flights) / len(df_flights)) * 100

plt.text(15, plt.ylim()[1] * 0.9, f'{percent_on_time:.1f}% flights within [-5, 15] min',
         horizontalalignment='left', color='green', fontsize=12)

# Set title and labels
plt.title('Distribution of Arrival Delays', fontsize=16)
plt.xlabel('Arrival Delay (minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.75)

# Use tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
````

````python
# Prompt B
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate to daily average arrival delay
daily_avg_delay = df_flights.groupby(df_flights['date'].dt.date)['arr_delay_min'].mean().reset_index()
daily_avg_delay['date'] = pd.to_datetime(daily_avg_delay['date'])

# Calculate 7-day rolling average
daily_avg_delay['rolling_avg_delay'] = daily_avg_delay['arr_delay_min'].rolling(window=7).mean()

# Create the line chart
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='arr_delay_min', data=daily_avg_delay, label='Daily Average Delay')
sns.lineplot(x='date', y='rolling_avg_delay', data=daily_avg_delay, color='orange', label='7-Day Rolling Average')

# Add axis labels, title, and grid
plt.title('Daily Average Arrival Delay with 7-Day Rolling Average', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Arrival Delay (minutes)', fontsize=12)
plt.grid(True)
plt.legend()

# Use tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
````

````python
#Prompt A
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate On-Time Rate by carrier
on_time_rate_by_carrier = df_flights.groupby('carrier')['on_time'].mean().reset_index()
on_time_rate_by_carrier['on_time_rate'] = on_time_rate_by_carrier['on_time'] * 100

# Sort by On-Time Rate
on_time_rate_by_carrier = on_time_rate_by_carrier.sort_values('on_time_rate', ascending=False)

# Create the bar chart
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='carrier', y='on_time_rate', data=on_time_rate_by_carrier, palette='viridis')

# Add percentage labels on bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

# Set title and labels
plt.title('On-Time Rate by Carrier', fontsize=16)
plt.xlabel('Carrier', fontsize=12)
plt.ylabel('On-Time Rate (%)', fontsize=12)
plt.ylim(0, 100) # Set y-axis limit to 100%

# Use tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
````
